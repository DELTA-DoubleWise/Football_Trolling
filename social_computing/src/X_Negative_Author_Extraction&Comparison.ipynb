```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from datetime import datetime,timedelta
import tqdm.notebook as tq
```

```python
df_p1s1 = pd.read_csv('../data/comments/player/sentiment/Sterling_MNC_sentiment.csv')
df_p1s2 = pd.read_csv('../data/comments/player/sentiment/Sterling_national_team_sentiment.csv')
df_p1s3 = pd.read_csv('../data/comments/player/sentiment/Sterling_premier_sentiment.csv')
df_p1s4 = pd.read_csv('../data/comments/player/sentiment/Sterling_soccer_sentiment.csv')
df_p1s5 = pd.read_csv('../data/comments/player/sentiment/Sterling_football_sentiment.csv')
```

```python
df_p1s4_neg = df_p1s4.loc[df_p1s4.sentiment_score<-0.05]

df_as = df_p1s4.groupby(['author'], as_index = False)['sentiment_score'].count()
df_as.rename(columns={'sentiment_score':'count'},inplace=True)

df_as = df_as.loc[df_as['count']>10]

df_as_neg = df_p1s4_neg.groupby(['author'])['sentiment_score'].count()

neg = []
for i in tq.tqdm(range(df_as.shape[0])):
    if df_as.iloc[i]['author'] in df_as_neg.index:
        neg.append(df_as_neg[df_as.iloc[i]['author']])
    else:
        neg.append(0)

df_as['neg_count'] = neg
df_as['neg_rate'] = df_as['neg_count']/df_as['count']
```

```python
neg_author_p1s4 = df_as.sort_values(by='neg_rate',ascending=False)[:50]
```

```python
df_p1s1_neg = df_p1s1.loc[df_p1s1.sentiment_score<-0.05]

df_as = df_p1s1.groupby(['author'], as_index = False)['sentiment_score'].count()
df_as.rename(columns={'sentiment_score':'count'},inplace=True)

df_as = df_as.loc[df_as['count']>10]

df_as_neg = df_p1s1_neg.groupby(['author'])['sentiment_score'].count()

neg = []
for i in tq.tqdm(range(df_as.shape[0])):
    if df_as.iloc[i]['author'] in df_as_neg.index:
        neg.append(df_as_neg[df_as.iloc[i]['author']])
    else:
        neg.append(0)

df_as['neg_count'] = neg
df_as['neg_rate'] = df_as['neg_count']/df_as['count']
```

```python
neg_author_p1s1 = df_as.sort_values(by='neg_rate',ascending=False)[:50]
```

```python
authors = set(neg_author_p1s1.author).intersection((neg_author_p1s4.author))
```
