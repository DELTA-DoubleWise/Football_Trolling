```python
team_SPI = team.T
team_SPI
```

```python
team_names = list(team_SPI.iloc[0])
team_SPI.drop(['team_code','team_name','end_season_rank','end_season_points'],axis=0,inplace=True)
team_SPI.reset_index(inplace=True)
team_SPI.drop(['index'],axis=1,inplace=True)

team_SPI.rename(columns=lambda x: team_names[x], inplace=True)
team_SPI.rename(index=lambda x:int(x)+1,inplace=True)
```

```python
fig, ax = plt.subplots(1, 1)

fig.set_size_inches(6.4, 3.2)
ax = plt.gca()
for team in team_SPI.columns:
    ax.plot(team_SPI.index,team_SPI[team],label=team)
ax.set_xlabel(r'Week')
ax.set_ylabel('SPI rating')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('../../Final Report/Assets/Team SPI.pgf', bbox_inches='tight')
```
