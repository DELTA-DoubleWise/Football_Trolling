```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/match_troll/12h/MNC_match_troll.csv')
```

```python
data
```

```python
data[['date','opponent','is_win','is_tie','is_loss','my_goal','opponent_goal']]
```

```python
plt.figure(figsize = (12,9))
plt.plot()
ax = plt.gca()

plt.grid(axis='y', alpha=0.75)
ax.plot(data['id'],data['high_ups_rate'],color='#0504aa',marker='o')
for i in range(data.shape[0]):
    plt.annotate(data.opponent[i], xy = (data.id[i], data.high_ups_rate[i]), xytext = (data.id[i]-0.01, data.high_ups_rate[i]+0.005))
ax2 = ax.twinx()
ax2.plot(data['id'],data['high_ups_count'],color='r',marker='o')


plt.xlabel('Opponent')
plt.ylabel('high_ups_rate')
plt.title('high ups comments rate of Manchester City subreddit')
plt.show()
```

```python
plt.figure(figsize = (12,9))
plt.plot()
ax = plt.gca()

plt.grid(axis='y', alpha=0.75)
ax.plot(data['id'],data['negative_ups_rate'],color='#0504aa',marker='o')
for i in range(data.shape[0]):
    plt.annotate(data.opponent[i], xy = (data.id[i], data.negative_ups_rate[i]), xytext = (data.id[i]-0.0001, data.negative_ups_rate[i]+0.0005))
plt.xlabel('Opponent')
plt.ylabel('negative_ups_rate')
plt.title('Negative ups comments rate of Manchester City subreddit')
plt.show()
```

```python
plt.figure(figsize = (12,9))
plt.plot()
ax = plt.gca()

plt.grid(axis='y', alpha=0.75)
ax.plot(data['id'],data['comments_count'],color='#0504aa',marker='o')
for i in range(data.shape[0]):
    plt.annotate(data.opponent[i], xy = (data.id[i], data.comments_count[i]), xytext = (data.id[i], data.comments_count[i]+1))
plt.xlabel('Opponent')
plt.ylabel('#comments')
plt.title('#comments of subreddit')
plt.show()
```

```python
sns.jointplot('is_big_six', 'negative_ups_rate', data, kind='scatter', color='seagreen')
```
